```r
library(edgeR)
library(scran)
library(scater)
library(stringr)
```

```r
# Load my rds files as sce obejcts

write.dir <- "../../../write/03_batch_expression/PICA0001-PICA0007/PICA0001-PICA0007_combined_annot_with_age_scvi.sce.rds"
# Get a list of all .rds files in the directory
rds_files <- list.files(write.dir, pattern = "*.rds", full.names = TRUE)
sce.list <- list()

for (file in rds_files) {
    cell.type <- str_split(basename(file), "_", n = 2)[[1]][2] # n = the max number of pieces to return
    cell.type <- sub("\\.rds$", "", cell.type)
    
    sce <- readRDS(file)
    sce.list[[cell.type]] <- sce
}
```
